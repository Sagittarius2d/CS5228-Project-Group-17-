```python
# Loading dataset from google drive
import os
from google.colab import drive

drive.mount('/content/drive')

train_path = '/content/drive/MyDrive/cs5228/cs5228_proj/data/train.csv'
test_path = '/content/drive/MyDrive/cs5228/cs5228_proj/data/test.csv'
auxiliary_path = '/content/drive/MyDrive/cs5228/cs5228_proj/auxiliary-data/'
auxiliary_mrt = auxiliary_path+'sg-train-stations.csv'
```

```python
train = pd.read_csv(train_path)
test = pd.read_csv(test_path)

aux_df = pd.read_csv(auxiliary_mrt)
```

```python
# compute distance between mrt stations and house location using latitude and longtitude
from geopy.distance import geodesic

def compute_distance(row, aux_df):
    address_lat, address_lng = row[['latitude','longitude']]
    min_distance = 1
    for stn_lat, stn_lng in zip(aux_df['lat'], aux_df['lng']):
        distance = geodesic((address_lat, address_lng), (stn_lat, stn_lng)).km
        if distance < min_distance:
            return 1
    return 0
```

```python
# Compute nearby stations in train set
merged_train_df = train.copy()
merged_train_df['nearby'] = merged_train_df.apply(compute_distance, aux_df=aux_df, axis=1)
```

```python
# double check data size after being processed
unique, count = np.unique(merged_df['nearby'], return_counts=True)
print(unique, np.sum(count))
```

```python
merged_train_df.to_csv('/content/drive/MyDrive/cs5228/cs5228_proj/merged_mrt_train.csv')
```

```python
# Compute nearby station in test set
merged_test_df = test.copy()
merged_test_df['nearby'] = merged_test_df.apply(compute_distance, aux_df=aux_df, axis=1)
```

```python
unique, count = np.unique(merged_test_df['nearby'], return_counts=True)
print(unique, np.sum(count))
print(count)
```
